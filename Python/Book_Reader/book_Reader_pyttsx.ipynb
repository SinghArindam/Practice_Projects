```python
filename = "As A Man Thinketh.pdf"
```

```python
book = open(f"books/{filename}",'rb')
pdf_Reader = PyPDF2.PdfReader(book)
no_of_pages = len(pdf_Reader.pages)
print(no_of_pages)
```

```python
page = pdf_Reader.pages[12]
text = page.extract_text()
print(text)
```

```python
speaker = pyttsx3.init()
# rate = speaker.getProperty('rate')
# speaker.setProperty('rate', rate-50)
# speaker.save_to_file(text , 'test.mp3')
# # speaker.say(text)
# speaker.runAndWait()
```

```python
entire_text = ""
for num in range(no_of_pages):
    page = pdf_Reader.pages[num]
    text = page.extract_text()
    entire_text += text
print(f"{filename}:\nentire text :\n",entire_text)
```

```python
filename_only = filename.split(".")[0]
print(filename_only)
```

```python
txt = open(f"books/{filename_only}.txt","w+")
txt.write(entire_text)
txt.close()
```

```python
rate = speaker.getProperty('rate')
speaker.setProperty('rate', rate-50)
speaker.save_to_file(entire_text , f"audios/{filename_only}.mp3")
# speaker.say(text)
speaker.runAndWait()
```
